```python
import numpy as np 
import PIL 
import matplotlib.pyplot as plt 
import os
import glob
import random 
```

```python
dataset_dir = 'signatures/'
```

```python
os.listdir(dataset_dir)
```


1. full_forg: Contains 1320 forgery signatures (24 forgeries for each of 55 writers).
2. full_org: Contains 1320 genuine signatures (24 genuines for each of 55 writers). 

```python

forged = {}

for idx in range(1, 56):
    files = glob.glob(f"{dataset_dir}/full_forg/forgeries_{idx}_*.*")
    forged[f"forged_{idx}"] = files

originals = {}
for idx in range(1, 56):
    files = glob.glob(f"{dataset_dir}/full_org/original_{idx}_*.*")
    originals[f"original_{idx}"] = files
```

```python
def read_image(path) :
    
    #read image in grayscale mode
    img = PIL.Image.open(path).convert('L')
    img_to_array = np.array(img)
    return img_to_array 
    
```

```python
# Function to plot pairs of images
def plot_image_pairs(image_pairs, num_cols=2):
    fig = plt.figure(figsize=(16, 8))
    rows = len(image_pairs)
    columns = 2

    for idx, (org, forg) in enumerate(image_pairs):
        # Adds a subplot at the idx position
        ax1 = fig.add_subplot(rows, columns, 2*idx + 1)
        ax2 = fig.add_subplot(rows, columns, 2*idx + 2)

        # showing image
        ax1.imshow(org)
        ax1.axis('off')
        ax1.set_title("Original")

        # showing image
        ax2.imshow(forg)
        ax2.axis('off')
        ax2.set_title("Forged")

    plt.show()
```

```python

samples = []
for idx in range(1,6): 
    seq = int( np.random.choice( np.arange(1,56), size =1) )
    orig_sample = random.sample(originals[f'original_{seq}'], 1)[0]
    forged_sample = random.sample(forged[f'forged_{seq}'], 1)[0]
    im1 = read_image(orig_sample)
    im2 = read_image(forged_sample)
    samples.append( ( im1, im2) )
```

```python
plot_image_pairs(samples)
```
